```python
x = np.linspace(-np.pi, np.pi, 50)
```

```python
y = x
X, Y = np.meshgrid(x, y)
z = np.cos(Y)/(1 + np.square(X))
plt.contour(X,Y,z)
```

```python
plt.contour(X,Y,z,15)
```

```python
cp = plt.contour(X,Y,z)
plt.colorbar(cp)
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.plot_surface(X,Y,z)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X,Y,z)
ax.view_init(0, 30)
plt.show()
```
